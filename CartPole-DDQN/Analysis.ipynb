```python
from DDQN import DDQNAgent

rewards = []

if __name__ == "__main__":
    agent = DDQNAgent("CartPole-v1")
    
    num_episodes = 400
    for episode in range(num_episodes):
        agent.train()
        if episode % 100 == 0:
            test_reward = agent.test()
            print(f"Episode {episode}, Test Reward: {test_reward}")
            rewards.append(test_reward)
    final_reward = agent.test()
    print(f"Final Test Reward: {final_reward}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(rewards)
plt.title('DDQN Training Progress')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.grid(True)
plt.show()
```
